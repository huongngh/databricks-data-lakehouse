## Transformation logic

```python
query = """
SELECT
    sls.order_number  AS order_number,
    pr.product_key  AS product_key,
    cu.customer_key AS customer_key,
    sls.order_date AS order_date,
    sls.ship_date  AS shipping_date,
    sls.due_date   AS due_date,
    sls.sales_amount    AS sales_amount,
    sls.quantity AS quantity,
    sls.price    AS price
FROM silver.crm_sales sls
LEFT JOIN gold.dim_products pr
    ON sls.product_key = pr.product_number
LEFT JOIN gold.dim_customers cu
    ON sls.customer_id = cu.customer_id;
"""

df = spark.sql(query)
```

```python
df.limit(10).display()
```

## Write gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
